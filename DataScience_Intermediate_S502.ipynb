## Auto-agrupación (Closterización)

Podemos detectar grupos automáticos a partir de un conjunto de características.

```python
import pandas as pd

iris = pd.read_csv("/content/iris.csv")

iris["x1"] = iris["sepal.length"]
iris["x2"] = iris["sepal.width"]
iris["x3"] = iris["petal.length"]
iris["x4"] = iris["petal.width"]

X = iris[["x1", "x2"]]

X
```

```python
import seaborn as sns

sns.scatterplot(X, x="x1", y="x2")
```

```python
from sklearn.cluster import KMeans

clu = KMeans(n_clusters=2) # K-grupos automáticos

clu.fit(X) # Sólo con las características

y = clu.predict(X) # Obtenemos una Y desconocida (etiquetas automáticas 0 -> K-1)

y
```

```python
iris["y1"] = y

iris[["x1", "x2", "y1"]]
```

```python
import seaborn as sns

sns.scatterplot(iris, x="x1", y="x2", hue="y1")
```

```python
from sklearn.cluster import KMeans

clu = KMeans(n_clusters=3) # K-grupos automáticos

clu.fit(X) # Sólo con las características

y = clu.predict(X) # Obtenemos una Y desconocida (etiquetas automáticas 0 -> K-1)

y
```

```python
iris["y2"] = y

iris[["x1", "x2", "y2"]]
```

```python
import seaborn as sns

sns.scatterplot(iris, x="x1", y="x2", hue="y2")
```

```python
import seaborn as sns

sns.scatterplot(iris, x="x1", y="x2", hue="y2")
```

## Reducción de la Dimensalidad

Un conjunto de características podría tener variables linealmente dependientes entre ellas, reduciendo la calidad de los datos y haciendo que los algoritmos de aprendizaje se confundan. Entre más variables más complejidad de aprendizaje.

Una estrategia es reducir el número de características principales de análisis, mediante el supuesto que queremos construir ejes normales estándarizados que representen a la mayoría de los datos en variables linealmente independientes. Es decir, si tenemos múltiples variables donde algunas son interdependientes, por ejemplo, en 100 caracteríticas, 20 se explican a través de otras, como por ejemplo, que el peso se explique a través de la altura. Entonces nos gustaría reducir el número de ejes de análisis para:

1. Poder visualizar mejor los datos y entenderlos
2. Poder hacer más eficientes y rápidos los aprendizajes

```python
import pandas as pd

titanic = pd.read_csv("/content/titanic.csv")

titanic["y1"] = titanic["Survived"]

titanic["x1"] = titanic["Pclass"]
titanic["x2"] = titanic["Sex"].map({ "male": 1, "female": -1 })
titanic["x3"] = titanic["SibSp"]
titanic["x4"] = titanic["Parch"]
titanic["x5"] = titanic["Fare"]
titanic["x6"] = titanic["Embarked"].map({ "S": 1, "C": 2, "Q": 3 })

titanic[["x1", "x2", "x3", "x4", "x5", "x6", "y1"]]
```

### PCA (*Principal Component Analysis | Análisis de Componentes Principales*)

Un componente se considera como un eje deducido a partir de otros ejes, por ejemplo, si tenemos los ejes $x_1, x_2, ..., x_d$, pordemos pensar en los compontes $C_1, C_2, ..., C_k$ tales que:

$C_1 = F_1(x_1, x_2, ..., x_d)$

$...$

$C_k = F_k(x_1, x_2, ..., x_d)$

El PCA es un método basado en la Covarianza entre los ejes que permite extraer los componentes principales.

$\overline{C} = PCA(\overline{X})$

Donde

$\overline{C} = C_1, C_2, ..., C_k$

$\overline{X} = x_1, x_2, ..., x_d$

Suponiendo que la $k << d$ la dimensión $k$ es mucho menor a la dimensión $d$ estaríamos reduciendo significativamente el espacio.

```python
X = titanic[["x1", "x3"]].dropna().copy()

X
```

```python
import seaborn as sns

sns.displot(X, x="x1", y="x3", hue=titanic["y1"])
```

```python
from sklearn.decomposition import PCA

C = PCA(n_components=2).fit_transform(X)

Xs = pd.DataFrame(C, columns=["c1", "c2"])

Xs
```

```python
import seaborn as sns

sns.displot(Xs, x="c1", y="c2", hue=titanic["y1"])
```

```python
from sklearn.svm import SVC

clf = SVC()

clf.fit(X, titanic["y1"])

clf.score(X, titanic["y1"])
```

```python
from sklearn.svm import SVC

clf = SVC()

clf.fit(Xs, titanic["y1"])

clf.score(Xs, titanic["y1"])
```

### Reducción de dimensionalidad en el IRIS

```python
import pandas as pd

iris = pd.read_csv("/content/iris.csv")

iris["x1"] = iris["sepal.length"]
iris["x2"] = iris["sepal.width"]
iris["x3"] = iris["petal.length"]
iris["x4"] = iris["petal.width"]
iris["y"] = iris["variety"].map({ "Setosa": 0, "Versicolor": 1, "Virginica": 2 })

X = iris[["x1", "x2", "x3", "x4"]]

Y = iris["y"]

Xs = PCA(n_components=2).fit_transform(X)

from sklearn.svm import SVC

clf1 = SVC()

clf1.fit(X, Y) # 150 * 4 = 600

score1 = clf1.score(X, Y)

clf2 = SVC()

clf2.fit(Xs, Y) # 150 * 2 = 300 (50% -> 2x)

score2 = clf2.score(Xs, Y)

score1, score2
```

```python
import seaborn as sns

sns.scatterplot(x=Xs[:,0], y=Xs[:,1], hue=Y)
```

```python
import pandas as pd

iris = pd.read_csv("/content/iris.csv")

iris["x1"] = iris["sepal.length"]
iris["x2"] = iris["sepal.width"]
iris["x3"] = iris["petal.length"]
iris["x4"] = iris["petal.width"]
iris["y"] = iris["variety"].map({ "Setosa": 0, "Versicolor": 1, "Virginica": 2 })

X = iris[["x1", "x2", "x3", "x4"]]

Y = iris["y"]

Xs = PCA(n_components=1).fit_transform(X)

from sklearn.neural_network import MLPClassifier

clf1 = MLPClassifier()

clf1.fit(X[["x2"]], Y) # 150 * 4 = 600

score1 = clf1.score(X[["x2"]], Y)

clf2 = MLPClassifier()

clf2.fit(Xs, Y) # 150 * 2 = 300 (50% -> 2x)

score2 = clf2.score(Xs, Y)

score1, score2
```
