# Repaso de Datasets - Unión (`join` y `merge`) y Agrupamiento (`groupby` y `aggregate`)

```python
import pandas as pd

X = pd.DataFrame({
    "B": ["R", "S", "T"],
    "C": [1.2, 3.4, 5.6]
})

X.index = pd.Series([1, 2, 3], name="A")

X
```

```python
Y = pd.DataFrame({
    "D": [7, 9, 23, 24, 35],
    "A": [1, 1, 3, 2, 2],
    "E": [4.3, 5.6, 4.1, 3.8, 1.9]
})

Y
```

```python
Z = Y.join(X, on="A").copy()

Z
```

```python
Z["F"] = Z["E"] / Z["C"]

Z
```

```python
R1 = Z[["B", "F"]]

R1
```

```python
R1.groupby("B").aggregate("mean")
```

```python
X.join(Y, on="A")
```

```python
L = pd.DataFrame({
    "A": [1, 5, 8, 9],
    "B": [22, 34, 56, 17],
    "G": [40_000, 34_000, 56_000, 17_000],
    "C": ["R", "S", "T", "W"],
    "D": ["RED", "BLUE", "RED", "GREEN"]
})

L
```

```python
M = pd.DataFrame({
    "F": [22, 54, 77, 85, 99, 23, 42],
    "A": [1, 1, 5, 5, 8, 9, 8],
    "G": [22_000, 35_000, 49_000, 17_000, 24_000, 19_000, 11_000],
    "H": [5, 6, 8, 8, 9, 10, 11]
})

M
```

```python
N = pd.DataFrame({
    "H": [1,    2,    3,   4,   5,   5, 6,   7,   8,   9,  10,  11],
    "I": [0.6, 0.7, 0.2, 0.3, 0.4, 0.9, 1, 0.4, 0.7, 0.6, 0.8, 0.9]
})

N
```

```python
M.merge(L, left_on="A", right_on="A", suffixes=(".1", ".2"))
```

```python
R2 = M.merge(L, left_on="A", right_on="A", suffixes=("_M", "_L")).merge(N, left_on="H", right_on="H")

R2
```

```python
R2[["C", "I"]].groupby("C").aggregate("mean") # count | sum | mean | std | var | min | max | ...
```

```python
R2[["D", "G"]].groupby("D").aggregate("count") # count | sum | mean | std | var | min | max | ...
```

```python
R2[["D", "G"]].groupby("D").aggregate("sum") # count | sum | mean | std | var | min | max | ...
```
