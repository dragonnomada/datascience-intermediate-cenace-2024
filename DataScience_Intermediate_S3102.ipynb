```python
import openml

data = openml.datasets.get_dataset(151).get_data(dataset_format="dataframe")[0]

data
```

```python
data["nswprice"].quantile(0.95)
```

```python
mask1 = data["nswprice"] >= data["nswprice"].quantile(0.05)
mask2 = data["nswprice"] <= data["nswprice"].quantile(0.95)
mask3 = data["vicprice"] >= data["vicprice"].quantile(0.05)
mask4 = data["vicprice"] <= data["vicprice"].quantile(0.95)
mask5 = data["vicdemand"] >= data["vicdemand"].quantile(0.05)
mask6 = data["vicdemand"] <= data["vicdemand"].quantile(0.95)
mask7 = data["vicdemand"] >= data["vicdemand"].quantile(0.05)
mask8 = data["vicdemand"] <= data["vicdemand"].quantile(0.95)

data1 = data[mask1 & mask2 & mask3 & mask4 & mask5 & mask6 & mask7 & mask8].copy()

data1
```

```python
sns.boxplot(data1[["nswprice", "vicprice", "nswdemand", "vicdemand"]])
```

```python
data1
```

```python
data1["nswprice"].mean() / data1["vicprice"].mean()
```

```python
data1["x1"] = data1["nswprice"]
data1["x2"] = data1["vicprice"] * (data1["nswprice"].mean() / data1["vicprice"].mean())

data1
```
