```python
import numpy as np

x1 = np.random.normal(70, 10, 100) # Normal > PROMEDIO | DESVIACIÓN ESTÁNDAR

# Intervalo de Confianza al 66%: (70 - 10, 70 + 10) | (60, 80)
# Intervalo de Confianza al 95%: (70 - 20, 70 + 20) | (50, 90)
# Intervalo de Confianza al 99.5%: (70 - 30, 70 + 30) | (40, 100)

x1
```

```python
import matplotlib.pyplot as plt

plt.hist(x1)
plt.show()
```

```python
plt.scatter(range(len(x1)), x1)
plt.ylim((0, 110))
plt.show()
```

```python
x2 = (x1 - x1.min()) / (x1.max() - x1.min()) # Normalización lineal

x2
```

```python
plt.scatter(range(len(x2)), x2)
plt.ylim((0, 1))
plt.show()
```

```python
x3 = (x1 - x1.mean()) / x1.std()

x3
```

```python
plt.scatter(range(len(x3)), x3)
plt.ylim((-3, 3))
plt.show()
```

```python
x4 = np.random.uniform(40, 100, 100)

x4
```

```python
plt.hist(x4)
plt.show()
```

```python
x5 = (x4 - x4.min()) / (x4.max() - x4.min())
x6 = (x4 - x4.mean()) / x4.std()
```

```python
plt.scatter(range(len(x4)), x4)
plt.show()
```

```python
plt.scatter(range(len(x5)), x5)
plt.show()
```

```python
plt.scatter(range(len(x6)), x6)
plt.show()
```
