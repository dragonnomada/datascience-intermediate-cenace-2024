## Repaso de Pandas

```python
import pandas as pd

s1 = pd.Series([1, 3, 5, 7])
# s1 = pd.serie([1, 3, 5, 7]) # ERROR
# s1 = pd.Serie([1, 3, 5, 7]) # ERROR

s1
```

```python
s2 = s1 ** 2 # Operador exponente aplicado a cada elemento (element-wise | elemento-por-elemento)
             # [1, 3, 5, 7] -> [1 ** 2, 3 ** 2, 5 ** 2, 7 ** 2] -> [1, 9, 25, 49]

s2
```

```python
s2 = s1.map(lambda x: x ** 2) # <serie>.map(T) | T: lambda x: <y = transform(x)>

s2
```

```python
s2 = s1.apply(lambda x: x ** 2) # <serie>.apply(T) | T: lambda x: <y = rule(x)>

s2
```

```python
s3 = (2 * s2 / s1) ** 0.5 + np.random.normal(0, 2, s1.count())
# s3 = (2 * [1, 9, 25, 49] / [1, 3, 5, 7]) ** 0.5 + [-0.35, 0.68, 0.01, -0.0005]
#    = ([2, 18, 50, 98] / [1, 3, 5, 7]) ** 0.5 + [-0.35, 0.68, 0.01, -0.0005]
#    = ([2, 6, 10, 14]) ** 0.5 + [-0.35, 0.68, 0.01, -0.0005]
#    = [1.41, ..., ..., ...] + [-0.35, 0.68, 0.01, -0.0005]
#    = [-2.06, 0.57, 5.23, 2.55]

s3
```

Repaso de Ciencia de Datos

```python
titanic.head()
```

```python
titanic2 = titanic[["Pclass", "Age", "SibSp", "Parch", "Sex", "Survived"]].copy()

titanic2["Age"] = titanic2["Age"].fillna(titanic2["Age"].mean())

titanic2 = titanic2.fillna(0)

titanic2["Sex"] = titanic2["Sex"].map({ "male": 1, "female": -1 })

titanic2["Age"] = titanic2["Age"].apply(int)

titanic2.head(5)
```

```python
X = titanic2[["Pclass", "Age", "SibSp", "Parch", "Sex"]].values
Y = titanic2["Survived"].values
```

```python
X[:3]
```

```python
Y[:3]
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()

clf.fit(X, Y)

clf.score(X, Y)
```

```python
clf.predict([
    [ 3, 22,  1,  0,  1],
    [ 1, 38,  1,  0, -1],
    [ 3, 26,  0,  0, -1],
    [ 1, 15, 1, 1, 1],
    [ 1, 15, 1, 1, -1],
    [ 1, 29, 1, 1, 1], # H 29 años -> MUERE
    [ 1, 29, 1, 1, -1], # M 29 años -> VIVE
])
```

```python
from sklearn.tree import plot_tree

plot_tree(clf)
```
